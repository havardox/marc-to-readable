```python
from typing import List
from collections import defaultdict

import pymarc
from pymarc import Record
```

```python
# Open the MARC file
records: List[Record] = pymarc.marcxml.parse_xml_to_array(
    "data/ERB_eestikeelne_raamat.xml"
)
```

```python
# Initialize dictionaries to store counts
field_counts = defaultdict(int)

# Loop through each record in the MARC file
for record in records:
    # Count data fields and subfields
    for field in record.get_fields():
        tag = field.tag
        if field.is_control_field():
            field_counts[tag] += 1
        else:
            for subfield in field.subfields:
                field_counts[f"{tag}${subfield.code}"] += 1

# Print the counts
print("Field Counts:")
for field in sorted(field_counts, key=field_counts.get, reverse=True):
    print(f"{field}:\t{field_counts[field]}")
```
